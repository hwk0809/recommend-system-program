```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D        #3d绘图必备，操！
import pandas as pd
from matplotlib.font_manager import FontProperties  # 字体管理器
import seaborn as sns
import matplotlib as mpl
```

```python
plt.ion()
x = list(range(6))
y = [num**2 for num in x]
plt.plot(x, x, lw='2', ls='-', marker='^')
plt.plot(x, y, lw='2', ls='-', marker='o')
plt.title('Some Curves', fontsize=20)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.legend(['y=x', 'y=x^2'], loc='upper left')
plt.xlim([-1, 6])
plt.ylim([-10, 30])
plt.grid()

# plt.annotate('Max', xy=(5, 25), )          FIGURE:图    AXES:图区   AXIS:轴
plt.annotate('Max', xy=(5, 25), xytext=(4.1, 27), arrowprops=dict(facecolor='black', width=1.5))
plt.savefig('D:/my figure.png')  # 保存图片，注意savefig命令必须在show命令之前
plt.show()  # 显示图形，在非交互绘图环境下需要调用show命令才能看到图像

#两张图之间不用show，则同一绘图
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.1f%%',shadow=True, startangle=0)# 初始旋转角度，从右边水平线开始
plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
plt.hist(x, 50, alpha=0.75)  # 50代表划分为50个区间，将输入数据按照值域分拆为50份，然后频率计数，用于一组数的频率记录
plt.xlabel('Smarts')
plt.ylabel('Frequency')
# 使用'$ $'来输入数学公式或符号
plt.text(60, 500, r'$\mu=100,\ \sigma=15$')    #text妙啊！
plt.xlim(40, 160)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
# ax=fig.add_subplot(111, projection='3d')
# Prepare arrays x, y, z
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()
plt.show()
```

```python
plt.subplot(2, 1, 1)  
plt.plot([1, 2, 4])
# 如果子图小于9个，可以直接用xyz编码，其中x表示行，y表示列，z表示子图的索引位置（上-下，左-右）
plt.subplot(223)  
plt.bar(['A', 'B', 'C'], [1, 2, 4])
# 利用函数式api绘图，子图切分方式非常灵活，后面的切分不受前面切分方式的影响，例如第一次切分为2行一列，这里仍可以切分为2行2列
plt.subplot(224)  
plt.pie([1, 3, 6], labels=['a', 'b', 'c'])
plt.show()
```

```python
# 用subplots一次切分好区域并返回包含所有子图的对象的矩阵axes。
fig, axes = plt.subplots(nrows=2, ncols=2)
# 使用axes[i, j]来调用每个子图对象，此方法可以实现更多精细的操作。
axes[0, 0].plot([1, 2, 4])
axes[0, 1].bar(['A', 'B', 'C'], [1, 2, 4])
axes[1, 0].pie([1, 3, 6], labels=['a', 'b', 'c'])
axes[1, 1].barh(['A', 'B', 'C'], [1, 2, 4])
plt.show()
```

```python
sns.set(style="darkgrid")

# 加载tips数据集
tips = sns.load_dataset("tips")

# 绘制图形，根据不同种类的三点设定图注
sns.relplot(x="total_bill",
y="tip", 
	hue="smoker",
	style="time", 
	data=tips )
plt.show()
```

```python
# 定义主题风格
sns.set(style="darkgrid")

# 加载tips
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill",
            y="tip",
            hue="size",
            palette="ch:r=-.5,l=.75",
            data=tips)
plt.show()
```

```python
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "regression")))

#加载数据集
tips = sns.load_dataset("tips")

# 使用regplot绘制
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.set(color_codes=True)

#加载数据集
tips = sns.load_dataset("tips")

#绘制树形图
sns.swarmplot(x="day",
              y="total_bill",
              hue="sex",
              data=tips)
plt.show()
```

```python
sns.set_theme(color_codes=True)
penguins = sns.load_dataset("penguins")

sns.pairplot(penguins)
g = sns.pairplot(penguins, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
plt.show()
```
